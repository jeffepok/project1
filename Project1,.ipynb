## Question 2 Billboard weekly music rank dataset

```python
import pandas as pd 
billboard = pd.read_csv('billboard.csv')
billboard.head()
```

```python
week_list = list(billboard.columns)[5:] #the list of week strings to be passed as id_vars parameter
week_list[0:5] # sample of how it looks like
```

```python
tag_list = list(billboard.columns)[0:5] #the list containing the string artiste,track, genre,date.entered,date.peaked 
#to be passed as value_vars parameter
tag_list
```

```python
billboard_new = pd.melt(billboard, id_vars = tag_list , value_vars = week_list, 
        var_name = 'week', value_name = 'rank') #pivot
```

```python
billboard_new['week'] = billboard_new['week'].str.extract('(\d+)').astype(int) #extract the numbers from the week
```

```python
billboard_new['date.entered'] = pd.to_datetime(billboard_new['date.entered']) # convert date.entered to datetime
billboard_new['date.peaked'] = pd.to_datetime(billboard_new['date.peaked']) # convert date.peaked to datetime
billboard_new['entered_to_peak'] = (billboard_new['date.peaked']-billboard_new['date.entered']).dt.days # create
#new column entered_to_peak
billboard_new.head() #final look of the dataFrame
```

```python
billboard_new.dtypes
```

```python
billboard_new[(billboard_new['rank'] == 1) & (billboard_new['entered_to_peak'] == 35) & (billboard_new['genre'] == 'Rock')]
#Who and what music ranked 1 in Rock after peaking 35 days from entering into the competition
```
